```python
#we are going to autmoate this.
def check_price():
    from bs4 import BeautifulSoup as bs
    import requests
    from urllib.request import urlopen as uReq
    import random
    import time
    import datetime
    import smtplib
    headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36" 
            }
    url="https://www.amazon.in/DUDEME-Programmer-Coding-Developer-T-Shirt/dp/B08SFHKHV3/ref=sr_1_1?crid=1PHEQS7XA9JNZ&keywords=data+%2Banalyst+%2Btshirt&qid=1676625350&sprefix=data+%2Banalyst+%2Btshirt%2Caps%2C310&sr=8-1"
    soup1=bs(page.content,"html.parser")
    soup2=bs(soup1.prettify(),"html.parser")
    title=soup2.find(id="productTitle")
    title.get_text()
    price=soup2.find(id="corePrice_desktop")
    price_span = price.find('span', {'class': 'a-offscreen'})
    price_text = price_span.text.strip()
    price_text
    star=soup2.find("span",{"class":"a-icon-alt"})
    star.text
    star=star.text.strip()
    name=title.get_text().strip()
    import datetime
    today=datetime.date.today()
    import csv
    #in the csv we want header and data.
    header=["Name","Price_text","date"]
    data=[name,price_text,today]
    with open("AmazonWebSraperDataset.csv","a+",newline='',encoding="UTF8") as f:
        writer=csv.writer(f)
    #writer.writerow(header)#wo dont need thee heafer anymore  
        writer.writerow(data)
    #if (price<300):
        #send_mail()
    
    
#we are going to put this on timer.
#whenever we want to check it we can do it, be it a minute, a day, sec.
    
```

```python
while(True):
    check_price()
    time.sleep(5)   #we are putting 5 second.
    
#every 5 second it is going to run this entire process.
#generally we dont do anything every 5 seconds.
#it is going to check the price every 5 seconds.
#if the amazon changes the price we will see it in our data.
#we could do anything or anyproduct.
#we have autmoated the task.
#https://www.youtube.com/watch?v=HiOtQMcI5wg&t=1506s link to the project.
#we can only run this a single time.
```
